```python
#median without built in function
def my_median(sample):
    n=len(sample)
    index=n//2
#sample with an odd number of observation
    if n%2:
        return sorted(sample)[index]
#sample with an even number of observation
    return sum(sorted(sample)[index-1:index+1])/2
```
